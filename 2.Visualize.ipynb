#### ANALYSIS ON CATEGORICAL AND DISCRETE FEATURE

#### Create a new dataframe for categorical features.

```python
categorical = []
for column in data.columns:
    if data[column].dtype == object:
        categorical.append(column)
        print(f"{column} : {data[column].unique()}")
        print("==============================")
```

```python
categorical  # all categorical features
```

```python
# Make a list of categorical use and subplot to plot multiple feature
categorical_col=data[['Gender','EducationBackground','MaritalStatus','BusinessTravelFrequency','OverTime','Attrition']]
plt.figure(figsize=(20,22))  #defining canvas size
plotnumber=1 #counter

for column in categorical_col:  # iteration of columns / acessing the columns from count 
    if plotnumber<=7:  # set the limit
        plt.subplot(3,2,plotnumber) #plotting 6 graphs (3-rows,2-columns) ,plotnumber is for count
        sns.countplot(x=categorical_col[column],hue=data.PerformanceRating) # Plotting count plots because the feature is categorical
        plt.xlabel(column,fontsize=20) # assigning name on x-axis & font size is 20
        plt.ylabel('PerformanceRating')
    plotnumber+=1  #Counter Increment
plt.tight_layout()
plt.show()
```

### Observations:

#### Relationship between Gender and Performance Rating
* More number of male & female employees performating 3 rating.
* performance rating 4 is less for male & female employees.

#### Relationship between Education Background & Performance Rating
* Education background from Life science and medical gives 3 as performance rating.
* Education background from Life science and medical give 2 as performance rating. 

#### Relationship between Marital Status & Performance Rating.
* Most of married employees are rated with 3 performance rating.
* Single & Divorced employees are also rated with 3 performance rating.

#### Relationship between Business Travel Frequency & Performance Rating
* Employees who travel rarely for business performance rated as 3.
* Employees who travel frequently & non travel also performance rated as 3.

#### Relationship between OverTime & Performance Rating
